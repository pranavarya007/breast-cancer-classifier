```python
#Split the data into independent X and dependent Y data sets
X = df.iloc[:,2:].values
Y = df.iloc[:,1].values
type(X)
```

```python
#split the data into 75% training data and 25% test data set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.25, random_state=0)
```

```python
#scale the data (feature scaling)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train
```

```python
#Create a function fot the models 
def models(X_test, Y_train):

    #Logistic regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state=0)
    log.fit(X_train, Y_train)

    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion='entropy', random_state=0)
    tree.fit(X_train, Y_train)

    #Random Forest
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
    forest.fit(X_train, Y_train)

    #printing the model accuracy
    print("Logistic Regression accuracy: ", log.score(X_train, Y_train))
    print("Decision tree accuracy: ", tree.score(X_train, Y_train))
    print("Random Forest accuracy: ", forest.score(X_train, Y_train))

    return log, tree, forest
```

```python
model =models( X_train, Y_train)
```

```python
#test model accuracy on test data on confusion matrix
from sklearn.metrics import confusion_matrix
for i in range(len(model)):
    print('Model ', i)    
    cm = confusion_matrix(Y_test, model[i].predict(X_test))  

    true_positive = cm[0][0]
    false_positive = cm[0][1]
    false_negative = cm[1][0]
    true_nagative = cm[1][1]
    print(cm)
    print('model accuracy is = ', (true_positive+true_nagative)/(true_nagative+true_positive+false_negative+false_positive))
    print()
```

```python
#another way of getting matrix of the modles
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
for i in range(len(model)):
    print('Model ', i) 
    print(classification_report(Y_test, model[i].predict(X_test)))
    print(accuracy_score(Y_test, model[i].predict(X_test))) 
    print()
```
